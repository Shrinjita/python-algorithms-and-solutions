1. Implement an XOR Logic Gate using a Multi-layer Perceptron Model

```python
import numpy as np
from sklearn.neural_network import MLPClassifier
import plotly.graph_objs as go
import plotly.subplots as sp

# XOR input and output
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

# Create MLP model
mlp = MLPClassifier(hidden_layer_sizes=(2,), max_iter=10000, activation='relu')
mlp.fit(X, y)

# Predictions
y_pred = mlp.predict(X)

# Create grids for visualization
fig = sp.make_subplots(rows=1, cols=1)
trace = go.Scatter(x=X[:, 0], y=X[:, 1], mode='markers',
                   marker=dict(size=10, color=y_pred, colorscale='Viridis'),
                   text=['Output: {}'.format(val) for val in y_pred])

fig.add_trace(trace)
fig.update_layout(title="XOR Logic Gate with MLP",
                  xaxis_title="Input 1", yaxis_title="Input 2")
fig.show()
```

2. Generate a Random Dataset and Classify it Using 6 Multi-Layer Perceptron Models

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Generate random dataset
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train multiple MLP models
for i in range(6):
    mlp = MLPClassifier(hidden_layer_sizes=(10 + i*2,), max_iter=1000, activation='relu', random_state=42)
    mlp.fit(X_train, y_train)
    y_pred = mlp.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"MLP Model {i+1} Accuracy: {accuracy:.4f}")
```

5. Implement an XOR Logic Gate using TensorFlow

```python
import tensorflow as tf
import numpy as np

# XOR input and output
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Define the model
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=2, activation='relu', input_dim=2))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=1000, verbose=0)

# Predict
predictions = model.predict(X)
print(np.round(predictions))
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the activation functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    e_x = np.exp(x - np.max(x))  # For numerical stability
    return e_x / e_x.sum(axis=0)

def tanh(x):
    return np.tanh(x)

# Generate input values
x = np.linspace(-10, 10, 400)

# Calculate the activation function outputs
relu_y = relu(x)
sigmoid_y = sigmoid(x)
tanh_y = tanh(x)

# Softmax is usually applied to a vector, so we need multiple vectors to show its effect
x_softmax = np.linspace(-10, 10, 400)
softmax_y = np.vstack([softmax(x_softmax - i) for i in range(3)])  # Three different softmax inputs

# Create subplots
fig = go.Figure()

# ReLU plot
fig.add_trace(go.Scatter(x=x, y=relu_y, mode='lines', name='ReLU'))

# Sigmoid plot
fig.add_trace(go.Scatter(x=x, y=sigmoid_y, mode='lines', name='Sigmoid'))

# Tanh plot
fig.add_trace(go.Scatter(x=x, y=tanh_y, mode='lines', name='Tanh'))

# Softmax plot
for i in range(softmax_y.shape[0]):
    fig.add_trace(go.Scatter(x=x_softmax, y=softmax_y[i], mode='lines', name=f'Softmax - Input {i}'))

# Update layout
fig.update_layout(title='Activation Functions',
                  xaxis_title='Input',
                  yaxis_title='Output',
                  legend_title='Activation Functions')

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the activation functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    e_x = np.exp(x - np.max(x))  # For numerical stability
    return e_x / e_x.sum(axis=0)

def tanh(x):
    return np.tanh(x)

# Create a range of values
x = np.linspace(-10, 10, 400)
x_softmax = np.linspace(-10, 10, 400).reshape(-1, 1)  # For softmax

# Compute activation function values
y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_softmax = softmax(x_softmax)
y_tanh = tanh(x)

# Plot ReLU
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.lineplot(x=x, y=y_relu, color='blue')
plt.title('ReLU Activation Function')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('ReLU(x)')

# Plot Sigmoid
plt.subplot(2, 2, 2)
sns.lineplot(x=x, y=y_sigmoid, color='green')
plt.title('Sigmoid Activation Function')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')

# Plot Softmax
plt.subplot(2, 2, 3)
sns.lineplot(x=x_softmax.flatten(), y=y_softmax.flatten(), color='red')
plt.title('Softmax Activation Function')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('Softmax(x)')

# Plot Tanh
plt.subplot(2, 2, 4)
sns.lineplot(x=x, y=y_tanh, color='purple')
plt.title('Tanh Activation Function')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('Tanh(x)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=10):
        self.weights = np.zeros(input_size + 1)  # +1 for bias term
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        return self.activation(summation)

    def train(self, training_inputs, labels):
        for _ in range(self.epochs):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                error = label - prediction
                self.weights[1:] += self.learning_rate * error * inputs
                self.weights[0] += self.learning_rate * error

# Training data for AND gate
training_inputs_and = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels_and = np.array([0, 0, 0, 1])

# Training data for OR gate
training_inputs_or = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels_or = np.array([0, 1, 1, 1])

# Create Perceptron instances for AND and OR gates
perceptron_and = Perceptron(input_size=2)
perceptron_and.train(training_inputs_and, labels_and)

perceptron_or = Perceptron(input_size=2)
perceptron_or.train(training_inputs_or, labels_or)

# Testing
print("AND Gate:")
for input in training_inputs_and:
    print(f"Input: {input} => Output: {perceptron_and.predict(input)}")

print("\nOR Gate:")
for input in training_inputs_or:
    print(f"Input: {input} => Output: {perceptron_or.predict(input)}")
```

```python
from sklearn.neural_network import MLPClassifier
import numpy as np

# Training data for XOR gate
training_inputs_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels_xor = np.array([0, 1, 1, 0])

# Create and train the neural network
mlp = MLPClassifier(hidden_layer_sizes=(2,), activation='relu', max_iter=10000, solver='adam', random_state=1)
mlp.fit(training_inputs_xor, labels_xor)

# Testing
print("XOR Gate:")
for input in training_inputs_xor:
    # Reshape input to be 2D (1 sample, 2 features)
    input_reshaped = input.reshape(1, -1)
    # Predict and print output
    print(f"Input: {input} => Output: {mlp.predict(input_reshaped)[0]}")
```
